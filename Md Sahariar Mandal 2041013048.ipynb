```python
import random 
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import collections
import seaborn as sns

```

```python
#Loop for values b/w 60 and 70
Population = [i for i in range(60,70)]
#Choosing 1000 values
[Population.append(random.choice(Population)) for i in range(1000)]
print(Population)
```

```python
#Pandas series
dataset=pd.Series(Population)
dataset
```

```python
#Mean of the dataset
MEAN=dataset.mean()
print("Mean of the dataset : ",MEAN)  
```

```python
#Standard deviation of the dataset
SD=dataset.std()
print("Standard deviation of the dataset is : ",SD)
```

```python
#Plotting histogram
Freq=collections.Counter(Population)
plt.hist(Population)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of 1000 random values between 60 and 70")
plt.show()
```

```python
Sample=[]
S_MEAN=[]
# Mean for each list of sample size 30
for i in range(30):
    for j in range(30):
        val = random.choice(Population)
        Sample.append(val)
    temp_mean = np.mean(Sample)
    S_MEAN.append(temp_mean)
print(S_MEAN)
```

```python
#Repetition for more samples
for i in range(500):
    for j in range(20):
        val = random.choice(Population)
        Sample.append(val)
    temp_mean = np.mean(Sample)
    S_MEAN.append(temp_mean)
print(S_MEAN)
```

```python
sns.displot(S_MEAN,kind='kde')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title("Density function of sampling mean")
plt.show()
```

```python
print("Sample mean : ",np.mean(S_MEAN))
```

```python
print("Standard deviation : ",np.std(S_MEAN))
```
